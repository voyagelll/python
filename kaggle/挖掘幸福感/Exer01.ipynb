```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from scipy.stats import skew 
```

```python
train = pd.read_csv('happiness_train_complete.csv', encoding='gbk')
test = pd.read_csv('happiness_test_complete.csv', encoding='gbk')
```

```python
train.head()
```

```python
df = pd.concat((train, test))
```

```python
train.corr()['happiness'].sort_values(ascending=False)
```

```python
del_cols = ['id', 'happiness', 'survey_time']
```

```python
df.drop(del_cols, axis=1, inplace=True)
```

```python
num_to_str = ['survey_type', 'province', 'city', 'county', 'gender', 'nationality', 'religion',
              'edu', 'edu_status', 'political', 'join_party','health', 'health_problem', 'depression', 'hukou',
             'hukou_loc', 'socialize', 'relax','learn', 'social_neighbor', 'social_friend', 'socia_outing', 'equity', 'class',
             'class_10_before', 'class_10_after', 'class_14', 'work_exper', 'work_status', 'work_yr', 'work_type',
             'work_manage', 'family_status', 's_edu', 's_political', 's_hukou', 's_work_exper', 's_work_status', 's_work_type',
             'f_edu', 'f_political', 'f_work_14', 'm_edu', 'm_political', 'm_work_14', 'status_peer', 'status_3_before',
             'view', 'neighbor_familiarity']
```

```python
for col in num_to_str:
    df[col] = df[col].fillna('None')
    df[col] = df[col].apply(str)
```

```python
year_cols = ['birth', 'edu_yr', 'work_yr', 'marital_1st', 'marital_now', 'f_birth', 'm_birth']
```

```python
for col in year_cols:
    df[col] = df[col].fillna('None')
    df[col] = df[col].apply(str)
```

```python
other = ['edu_other', 'property_other', 'invest_other']
```

```python
for col in other:
    df[col] = df[col].fillna('None')
    df[col] = df[col].apply(lambda x: x if x=='None' else '1')
```

```python
num = ['religion_freq', 'income', 'floor_area', 'height_cm', 'weight_jin','family_income', 'family_m', 'house', 'car',
       'son', 'daughter', 'marital', 's_income', 'inc_ability']
```

```python
for col in num:
    df[col] = df[col].apply(lambda x: 0 if x<0 else x)
    df[col] = df[col].fillna(0)
    if df[col].skew() > 1:
        df[col] = df[col].apply(np.log1p)
```

```python
'property_0'- 'property_8'
'media_1' - 'media_6'
'leisure_1' - 'leisure_12'
'insur_1 - insur_4'
'invest_1' - 'invest_8'
'trust_1'- 'trust_13'
'public_service_1' - 'public_service_9'
```

```python
dum_cols = []
for col in df.columns:
    if ('property' in col) or 'media' in col or 'leisure' in col or 'insur' in col or 'invest' in col or 'trust' in col or 'public' in col:
        dum_cols.append(col)
```

```python
df = pd.get_dummies(df)
```

```python
df = df.fillna(df.mean())
```

```python
x_train = df[:len(train)]
y = train['happiness']
x_test = df[len(train):]
```

## model

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, LassoCV, Lasso 
from sklearn.model_selection import cross_val_score 
import lightgbm as lgb

def acc(model):
    score = cross_val_score(model, x_train, y, scoring='accuracy', cv=5)
    return score 
```

```python
lgr = lgb.LGBMClassifier(objective='Regressor',
                         num_leaves=20,
                         min_data_in_leaf=20,
                         max_depth=6,
                         learning_rate=0.01,
                         min_child_samples=30,
                         feature_fraction=0.8,
                         bagging_freq=1,
                         bagging_fraction=0.8,
                         bagging_seed=11,
#                          metric='mse',
                         lambda_l1=0.1,
                         n_estimators=1000,
                       )
score = acc(lgr)
```

```python
print(score)
```

```python
lgr = lgr.fit(x_train, y)
pred = lgr.predict(x_test)
```

```python
sub = pd.DataFrame()
sub[id] = test.id 
sub['happiness'] = pred
sub.to_csv('submission_2020_03_07.csv', index=False)
```
