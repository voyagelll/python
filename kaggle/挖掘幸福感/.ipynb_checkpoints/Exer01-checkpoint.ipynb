```python
import pandas as pd 
pd.set_option('display.max_columns',200)
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

from  datetime import datetime 
import re 

train = pd.read_csv('happiness_train_complete.csv', encoding='gbk')
test = pd.read_csv('happiness_test_complete.csv', encoding='gbk')
# train.head()
```

```python
df = pd.concat((train, test))
```

```python
df.drop(['id', 'happiness'], axis=1, inplace=True)
```

```python
df['survey_month'] = df['survey_time'].apply(lambda x: datetime.strptime(x, '%Y/%m/%d %H:%M').month)
df['survey_month'] = df['survey_time'].apply(lambda x: datetime.strptime(x, '%Y/%m/%d %H:%M').day)
```

```python
df.loc[df['nationality']==-8, 'nationality'] = 8
```

```python
df.loc[df['religion']==-8, 'religion'] = 1
```

```python
df['religion_freq'] = df['religion_freq'].replace(-8, 1)
```

```python
df['edu'] = df['edu'].replace(-8, 14)
```

```python
df.edu_other = df.edu_other.fillna('None')
```
