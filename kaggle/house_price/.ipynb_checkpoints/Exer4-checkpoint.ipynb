## Modeling

```python
from sklearn.grid_search import GridSearchCV 
from sklearn.model_selection import train_test_split, cross_val_score

import lightgbm as lgb
from xgboost import XGBRegressor
```

```python
df_copy = df.copy()
df_y = np.log1p(df_copy['SalePrice'])
df_copy.drop(['Id', 'SalePrice'], axis=1, inplace=True)
df_x = pd.get_dummies(df_copy)
x = df_x[:len(train)-2]
y = df_y[:len(train)-2]
test_x = df_x[len(train)-2:]

train_x, valid_x, train_y, valid_y = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
paramaters = {
    'max_depth':[4],
    'num_leaves':[3],
    'feature_fraction': [0.2],
    'cat_smooth': [1],
    'bagging_fraction':[0.9],
    'bagging_freq': [3],
    'cat_smooth':[0]
    
}

gbm = lgb.LGBMRegressor(
                    objective='regression',
                    learning_rate = 0.01,
                    n_estimators=5000,
                    verbose = -1
)

gscv = GridSearchCV(gbm,
                    param_grid=paramaters,
                    cv=5,
                    n_jobs=-1
                   )
gscv.fit(x, y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
np.sqrt(np.sum(np.power((gscv.best_estimator_.predict(valid_x) - valid_y),2))/len(valid_y))
```

```python
params = {
    max_depth=[3,4,5]
}
xgboost = XGBRegressor(learning_rate=0.01,
                       n_estimators=3460,
                       min_child_weight=0,
                       gamma = 0,
                       subsample=0.7,
                       colsample_bytree=0.7,
                       objective='reg:linear',
                       nthread=-1,
                       scale_pos_weight=1,
                       seed=27,
                       reg_alpha=0.00006,
                       random_state=42)
gscv_xg = GridSearchCV(xgboost,
                       cv=5,
                       param_grid=params
                       n_jobs=-1)
gscv_xg.fit(x, y)
```

```python
pred = gscv.predict(test_x)
submission = pd.DataFrame({'Id':test.Id, 'SalePrice':np.expm1(pred)})
submission.to_csv('submission_2020_02_19.csv', index=False)
```

```python
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from lightgbm import LGBMRegressor
```

```python
y = np.log1p(df['SalePrice'][:len(train)])
df.drop(['Id', 'SalePrice'], axis=1, inplace=True)
df = pd.get_dummies(df)
X_train = df[:len(train)]
X_test = df[len(train):]
```

```python
kfolds = KFold(n_splits=10, shuffle=True, random_state=42)

def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X_train, y, scoring='neg_mean_squared_error', cv=5))
    return rmse
    
lightgbm = LGBMRegressor(objective='regression',
                        num_levels=4,
                        learning_rate=0.01,
                        n_estimators=5000,
                        bagging_freq=5,
                        bagging_seed=7,
                        feature_fraction=0.2,
                        feature_fraction_seed=7,
                        verbose=-1)

score = rmse_cv(lightgbm)
print(score.mean(), score.std())
```

```python
gscv.best_params_
```

```python
lightgbm = lightgbm.fit(X_train, y)
pred = lightgbm.predict(X_test)
pred
```

```python
submission = pd.DataFrame({'Id':test.Id, 'SalePrice':np.expm1(pred)})
submission.to_csv('submission_2020_02_16.csv', index=False)
```
