工作原理：存在一个样本数据集合，也称作训练样本，并且每个样本都存在标签。输入一个新的数据后，将新数据的每个特征与样本中对应特征想比较，然后提取样本中k 个最相似的数据出现次数最多的分类，作为新数据的类别。

```python
import numpy as np 
import operator

# KNN 算法
def KNN(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
#     print('dataSet:', dataSet)
#     print('dataSetSize: ', dataSetSize)
#     print('inX:', inX)
#     print('diffMat: ', np.tile(inX, (dataSetSize, 1)))
    diffMat = np.tile(inX, (dataSetSize, 1)) - dataSet 
    sqDiffMat = diffMat ** 2 
#     print(sqDiffMat)
    sqDistances = sqDiffMat.sum(axis=1)
#     print(sqDistances)
    distances = sqDistances ** 0.5 
#     print(distances)
    sortedDistIndicies = distances.argsort()
    print(sortedDistIndicies)
    classCount = {}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1 
#     print(classCount)
    return sorted(classCount)[0]
```

```python
# test
dataSet = [[0,0],[0,0.1],[0,0.2],[1,1],[1,1.1],[1,1.2]]
dataSet = np.array(dataSet)
labels = ['A', 'A', 'B', 'B']

inX = np.array([0.5,0.5])
KNN(inX, dataSet, labels, 3)
```

```python
# 将文件内容转为矩阵
def file2matrix(filename):
    with open(filename, 'r') as f:
        content = f.readlines()
        rows = len(content)
        returnMat = np.zeros((rows, 3))
        labels = []
        index = 0
        for row in content:
            srow = row.strip().split('\t')
            returnMat[index,:] = srow[:3]
            labels.append(int(srow[-1]))
            index += 1
        return np.array(returnMat), np.array(labels)
    
returnMat, labels = file2matrix('datingTestSet2.txt')
# labels.map({'largeDoses':1, 'smallDoses':2, 'didntLike':3})
# returnMat
# labels
```

```python
# 特征归一化
def autoNorm(returnMat):
    minVals = returnMat.min(0)
    maxVals = returnMat.max(0)
    ranges = maxVals - minVals 

    normDataSet = returnMat - np.tile(minVals, (len(returnMat),1))
    normDataSet = normDataSet / np.tile(ranges, (len(returnMat), 1))
    return normDataSet, ranges, minVals
```

```python
# 测试代码
def test():
    ratio = 0.1 
    dataMat, labels = file2matrix('datingTestSet2.txt')
    normMat, ranges, minVals = autoNorm(dataMat)
    m = normMat.shape[0]
    testVecs = int(m * ratio)
    errorCount = 0 

    normMat[testVecs:m,:]
    for i in range(testVecs):
        result = KNN(normMat[i,:], normMat[testVecs:m, :], labels[testVecs:], 2)
#         print("The prediction is :", result, "The real answer is:", labels[i])
        if result!=labels[i] : errorCount += 1
    print("The total error rate is :", errorCount / testVecs)


test()
```

```python
import os 


def img2vec(filepath):
    os.chdir(filepath)
    files = os.listdir()
    vecs = []
    labels = []
    print(files)
    for file in files:
        if file is not None:
            vec = []
            labels.append(int(file[0]))
            with open(file, 'r', encoding='utf-8', errors='ignore') as f:
                print(file)
                content = f.read()
                content = content.replace('\n', '')
                for i in content:
                    vec.append(int(i))
            vecs.append(vec)
    return np.array(vecs), np.array(labels)
```

```python
filepath_training = 'C:\\Users\\Voyager\\Desktop\\machine_learning\\机器学习实战\K邻近\\trainingDigits'
training_vecs, training_labels = img2vec(filepath_training)
filepath_test = 'C:\\Users\\Voyager\\Desktop\\machine_learning\\机器学习实战\\K邻近\\testDigits'
test_vecs, test_labels = img2vec(filepath_test)
```
